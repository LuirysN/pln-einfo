```python
import transformers
from transformers import pipeline
```

```python
#MODELO BERTimbau
mascarar = pipeline("fill-mask", model="FacebookAI/xlm-roberta-base")
```

```python
# Tag <mask> é usada para mascarar aquilo que queres identificar
texto = mascarar("Rei está para Rainha assim como Homem está para <mask>")
for x in range(len(texto)):
  print(texto[x])
```

```python
texto = mascarar(["Existe uma chance do copo cair no [MASK]","Brasília é a capital do [MASK]"])
for x in range(len(texto)):
  print(texto[x])
```
