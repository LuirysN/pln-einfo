```python
from transformers import pipeline

# Carregar pipeline de análise de sentimentos (português ou multilinguístico)
sentiment_analysis = pipeline("sentiment-analysis", model="lxyuan/distilbert-base-multilingual-cased-sentiments-student")

# Exemplo de textos
texts = [
    "Este novo produto é incrível! Eu adorei.",
    "A equipe da SEFAZ é fantástica",
    "O serviço foi terrível e não pretendo voltar.",
    "Foi razoável, nem ótimo nem ruim.",
    "Oh, que companhia aérea fantástica! Conseguiram me enviar para um novo destino, \
    enquanto minha mala foi para outra aventura completamente diferente. Que surpresa!"
]

# Aplicar análise de sentimentos
results = sentiment_analysis(texts)

# Exibir os resultados
for text, result in zip(texts, results):
    print(f"Texto: {text}\nSentimento: {result['label']} (Confiança: {result['score']:.2f})\n")

```
