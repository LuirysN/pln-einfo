```python
import transformers
from transformers import pipeline
```

```python
gerador = pipeline("text-generation", model="pierreguillou/gpt2-small-portuguese")
```

```python
#no objeto, informa o texto, o tamanho e o do_sample
# Quando do_sample é definido como False, a geração de texto pode seguir uma abordagem determinística,
# onde o modelo escolhe sempre a palavra ou token com a maior probabilidade em cada passo.
# Isso resulta em sequências de texto mais determinísticas e previsíveis.
texto = "Em sentido estrito, ciência refere-se ao sistema de adquirir conhecimento baseado no método científico."
resultado = gerador(texto, max_length=1000, do_sample=True, truncation=False)
print(resultado)
```
