###Análise de Sentimentos com VADER


O VADER (Valence Aware Dictionary and sEntiment Reasoner) é uma ferramenta de análise de sentimentos especialmente projetada para lidar com textos em inglês. Sua abordagem é baseada em um dicionário de palavras com pontuações associadas que indicam a polaridade (positiva, negativa ou neutra) e a intensidade do sentimento expresso por essas palavras. O VADER retorna não apenas a classificação da sentença (positiva, negativa ou neutra) mas também uma pontuação de polaridade composta, indicando a intensidade do sentimento.
Há duas formas:
 - Polarity Score
 - Textblob

```python
#Vader só funciona em inglês
import nltk
nltk.download("vader_lexicon")
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
mas = SentimentIntensityAnalyzer()
```

```python
#Polarity Score
#a = 'I hate this movie'
#a = ':('
#a = 'It is true that this movie is good :('
a = 'Oh, what a fantastic airline! They managed to send me to a whole new \
#exciting destination, while my suitcase embarked on its own adventure to a completely \
#different place. Such a delightful surprise!'
x = mas.polarity_scores(a)
print(x)
```

Compound varia de -1 (muito negativo) a +1 (muito positivo).

```python
print(type(x))
```

```python
print(x['compound'])
```

```python
#textblob
#Retorna a polaridade [-1,1] e a subjetividade [0,1]
from textblob import TextBlob
#teste = TextBlob('I love Ibura')
#teste = TextBlob(':)')
teste = TextBlob("Oh, what a fantastic airline! They managed to send me to a whole new \
exciting destination, while my suitcase embarked on its own adventure to a completely \
different place. Such a delightful surprise")
print(teste.sentiment)
```
