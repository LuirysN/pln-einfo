```python
# **Passo 1: Configuração do Ambiente e Upload do Áudio**
from google.colab import files
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import widgets

# Botão para enviar áudio
print("Envie o arquivo de áudio para análise:")
uploaded = files.upload()
audio_path = list(uploaded.keys())[0]  # Obtém o nome do arquivo enviado
```

```python
from transformers import pipeline

# **Reconhecimento de fala com Whisper**
def speech_to_text_whisper(audio_path):
    """
    Usa o modelo Whisper para converter áudio em texto transcrito.
    Retorna o texto transcrito.
    """
    print("Iniciando reconhecimento de fala com Whisper...")

    # Carregar o pipeline do modelo Whisper
    transcriber = pipeline("automatic-speech-recognition", model="openai/whisper-large")

    # Realizar a transcrição
    transcription = transcriber(audio_path)["text"]

    print("Texto Transcrito (Whisper):", transcription)
    return transcription

# **Chamando as funções de reconhecimento de fala**
print("Executando transcrições com diferentes modelos...")

# Caminho para o arquivo de áudio
# Use o mesmo arquivo processado anteriormente
transcription_whisper = speech_to_text_whisper(audio_path)


# **Resumo das transcrições**
print("\n--- Resumo das Transcrições ---")
print(f"Whisper: {transcription_whisper}")

```

```python
# Modelo Baseado em Aprendizagem de Máquina
# Instalar a biblioteca gTTS (Google Text-to-Speech)
!pip install gtts
# Importar bibliotecas necessárias
from gtts import gTTS  # Google TTS para conversão de texto em fala
from IPython.display import Audio  # Para reproduzir o áudio no Jupyter Notebook

# Definir o texto que será sintetizado
text = "Olá, boa noite. Os ingredientes da Pizza de Calabresa são molho de tomate, queijo e linguiça calabresa."

# Criar objeto TTS (define a língua do texto como português)
tts = gTTS(text, lang="pt")

# Salvar o áudio gerado em um arquivo
output_file = "output_ml.mp3"  # Nome do arquivo de saída
tts.save(output_file)  # Salva o áudio gerado como MP3

print(f"Áudio gerado e salvo como '{output_file}'!")
from IPython.display import Audio
Audio(output_file)
```

```python
# Importar bibliotecas necessárias
from transformers import AutoProcessor, VitsModel
import torch
import torchaudio
from scipy.io.wavfile import write

# Carregar processador e modelo para TTS em Português Brasileiro
processor = AutoProcessor.from_pretrained("facebook/mms-tts-por")
model = VitsModel.from_pretrained("facebook/mms-tts-por")

# Texto de entrada (modifique conforme necessário)
text = "Olá, boa noite. Os ingredientes da Pizza de Calabresa são molho de tomate, queijo e linguiça calabresa."

# Processar o texto para gerar os tokens necessários para o modelo
inputs = processor(text, return_tensors="pt")

# Gerar o áudio com o modelo (sem cálculos de gradiente para economizar memória)
with torch.no_grad():
    waveform = model(**inputs).waveform

# Converter o tensor PyTorch para um formato que pode ser salvo
waveform = waveform.cpu().numpy().squeeze()

# Definir taxa de amostragem (sampling rate)
sample_rate = 16000  # A Hugging Face usa 16 kHz por padrão

# Salvar o áudio em um arquivo WAV
output_file = "output_pt_br.wav"
write(output_file, sample_rate, waveform)

print(f"Áudio gerado e salvo como '{output_file}'!")

# Reproduzir áudio no Jupyter Notebook
from IPython.display import Audio
Audio(output_file)
```

```python
# Importar bibliotecas necessárias
from transformers import AutoTokenizer, VitsModel
import torch
from scipy.io.wavfile import write

# Carregar tokenizer e modelo para TTS em Português Brasileiro
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-por")
model = VitsModel.from_pretrained("facebook/mms-tts-por")

# Texto de entrada
text = "Olá, boa noite. Os ingredientes da Pizza de Calabresa são molho de tomate, queijo e linguiça calabresa."

# Tokenizar o texto
inputs = tokenizer(text, return_tensors="pt")

# Gerar o áudio com o modelo
with torch.no_grad():
    waveform = model(**inputs).waveform

# Processar o waveform
waveform = waveform.cpu().numpy().squeeze()
sample_rate = 16000

# Salvar o áudio
output_file = "output_pt_br.wav"
write(output_file, sample_rate, waveform)

print(f"Áudio gerado e salvo como '{output_file}'!")

# Reproduzir no notebook
from IPython.display import Audio
Audio(output_file)
```
