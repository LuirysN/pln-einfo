LeiA (Lexicon for Informal Portuguese Language Sentiment Analysis) é uma ferramenta de análise de sentimentos baseada em léxico, desenvolvida para processar textos informais em português, especialmente em redes sociais.

Ela segue o mesmo princípio do VADER, mas com:

- léxico adaptado ao português brasileiro;

- tratamento de gírias, negações e emojis;

- regras ajustadas à morfologia e estrutura da língua portuguesa.

```python
from LeIA import SentimentIntensityAnalyzer

s = SentimentIntensityAnalyzer()

# Análise de texto simples
print(s.polarity_scores('Eu estou feliz'))

# Análise de texto com emoji :)
print(s.polarity_scores('Eu estou feliz :)'))

# Análise de texto com negação
print(s.polarity_scores('Eu não estou feliz'))

# Análise de texto com negação maiúscula
print(s.polarity_scores('Eu NÃO estou feliz'))

# Análise de texto com negação maiúscula e exclamações
print(s.polarity_scores('Eu NÃO estou feliz!!!!!'))

# Análise de texto com negação maiúscula, exclamações e emojis
print(s.polarity_scores('Eu não estou feliz!!!!!'))

# Análise de texto com negação maiúscula, exclamações e emojis
print(s.polarity_scores('Eu não estou feliz não!!!!!'))

# Análise de texto com negação maiúscula, exclamações e emojis
print(s.polarity_scores('Não, é verdade que eu NÃO estou feliz'))

# Análise de texto com negação maiúscula, exclamações e emojis
print(s.polarity_scores('Não é verdade que eu NÃO estou feliz'))

# Análise de texto com negação maiúscula, exclamações e emojis
print(s.polarity_scores('Não é verdade que eu NÃO estou feliz'))

# Análise de texto com negação maiúscula, exclamações e emojis
print(s.polarity_scores('Não é verdade que eu NÃO estou feliz!!!!'))

```

```python
# Análise de texto simples
print(s.polarity_scores('Eu estou pouco feliz!!!!'))

print(s.polarity_scores('Eu estou muito feliz!!!!'))
```
