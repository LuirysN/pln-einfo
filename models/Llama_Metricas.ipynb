### MÉTRICAS

**1**. BLEU (Bilingual Evaluation Understudy)
Compara o texto gerado com textos de referência usando correspondência de n-gramas.

Interpretação do valor do BLEU
Escala: O BLEU varia de 0 a 1:
- 1: Texto gerado é idêntico ao texto de referência.
- 0: Texto gerado não tem n-gramas em comum com o texto de referência.
- Melhor valor: Quanto mais próximo de 1, melhor o texto gerado, pois isso indica maior similaridade com o de referência.
- Limitação: BLEU não considera sinônimos nem a qualidade gramatical do texto.


Vamos comparar dois textos gerados com o texto de referência e observar os resultados:

- Texto 1 tem alta similaridade porque mantém a estrutura e as palavras-chave.
- Texto 2 possui termos importantes, mas a estrutura é mais diferente, resultando em um score BLEU menor.

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

# Texto de referência
reference_text = "A inteligência artificial tem impactos significativos na sociedade."

# Dois textos gerados para comparação
generated_text_1 = "A inteligência artificial impacta bastante a sociedade moderna."
generated_text_2 = "Impactos significativos da IA podem transformar a sociedade."

# Calcular BLEU para os dois textos gerados
#smoothing_function: Aplica um método de suavização para evitar que n-gramas ausentes resultem em um score zero (útil para pequenos textos ou n-gramas maiores).
score_1 = sentence_bleu([reference_text.split()], generated_text_1.split(), smoothing_function=SmoothingFunction().method1)
score_2 = sentence_bleu([reference_text.split()], generated_text_2.split(), smoothing_function=SmoothingFunction().method1)

print("BLEU Score (Texto 1):", score_1)
print("BLEU Score (Texto 2):", score_2)
```

ROUGE (Recall-Oriented Understudy for Gisting Evaluation)
Avalia a sobreposição de palavras e frases.

Tipos de ROUGE
- ROUGE-1: Mede correspondência de palavras únicas (unigramas).
- ROUGE-2: Mede correspondência de pares de palavras (bigramas).
- ROUGE-L: Mede a sequência mais longa de palavras correspondentes entre os textos (sequência comum mais longa).

Como interpretar os valores:
- Valores de ROUGE próximos de 1 indicam alta similaridade.
- ROUGE-1 alto significa que as palavras do texto gerado estão presentes no de referência.
- ROUGE-L alto indica que a estrutura geral e a ordem das palavras são semelhantes.

Vamos comparar dois textos gerados:
- O ROUGE-1 para o Texto 1 será maior porque mais palavras coincidem diretamente.
- O ROUGE-L avalia sequências mais longas, favorecendo textos com estrutura mais próxima do texto de referência

```python
from rouge_score import rouge_scorer

# Dois textos gerados para comparação
generated_text_1 = "A inteligência artificial impacta bastante a sociedade moderna."
generated_text_2 = "Impactos significativos da IA podem transformar a sociedade."
reference_text = "A inteligência artificial tem impactos significativos na sociedade."

# Inicializar o calculador de ROUGE
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

# Calcular ROUGE para os textos
scores_1 = scorer.score(reference_text, generated_text_1)
scores_2 = scorer.score(reference_text, generated_text_2)

print("ROUGE Scores (Texto 1):", scores_1)
print("ROUGE Scores (Texto 2):", scores_2)

```

Perplexity
Indica a fluência do texto gerado.

Interpretação
- Perplexidade Baixa: O texto é fluente e segue os padrões de linguagem aprendidos pelo modelo.
- Perplexidade Alta: O texto é menos fluente ou contém padrões inesperados, sugerindo que o modelo teve dificuldade em prever a sequência.

Vamos calcular a perplexidade para dois textos:

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Recarregar tokenizer e modelo GPT2 para perplexidade
perplexity_tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
perplexity_model = GPT2LMHeadModel.from_pretrained("gpt2")

# Dois textos gerados para comparação
generated_text_1 = "A inteligência artificial impacta bastante a sociedade moderna."
generated_text_2 = "Inteligencia IA artificial impacto modernidade sociedade."

# Calcular perplexidade para os dois textos
input_ids_1 = perplexity_tokenizer(generated_text_1, return_tensors="pt").input_ids
input_ids_2 = perplexity_tokenizer(generated_text_2, return_tensors="pt").input_ids

#O modelo calcula a perda (loss) para os textos fornecidos.
#A perda representa o erro do modelo ao prever a sequência de tokens.
loss_1 = perplexity_model(input_ids_1, labels=input_ids_1).loss
loss_2 = perplexity_model(input_ids_2, labels=input_ids_2).loss

#A perplexidade é derivada da exponenciação da perda:
#Perplexidade = 𝑒^(Loss)
perplexity_1 = torch.exp(loss_1)
perplexity_2 = torch.exp(loss_2)

print("Perplexity (Texto 1):", perplexity_1.item())
print("Perplexity (Texto 2):", perplexity_2.item())
```

Cosine Similarity (Semântica)
Usa embeddings para medir a similaridade semântica.

Como interpretar os resultados?
- Valores entre 0 e 1: Indicam que os vetores estão próximos no espaço, ou seja, são semântica ou contextualmente semelhantes.
 - 1: Vetores estão perfeitamente alinhados (máxima similaridade).
 - 0.7: Alta similaridade, mas com diferenças perceptíveis.
 - 0.3: Baixa similaridade.

- Valor 0: Indica que os vetores são ortogonais, ou seja, não compartilham nenhuma direção ou semelhança.
- Valores entre -1 e 0: Indicam que os vetores estão em direções opostas, ou seja, são semântica ou contextualmente opostos.
 -1: Vetores completamente opostos.

Vamos comparar dois textos gerados:

```python
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer

# Carregar modelo de embeddings
embedder = SentenceTransformer('all-MiniLM-L6-v2')

reference_text = "A tecnologia moderna altera significativamente nossa sociedade."

# Dois textos gerados para comparação
generated_text_1 = "A inteligência artificial impacta bastante a sociedade moderna."
generated_text_2 = "A tecnologia moderna altera significativamente nossa sociedade."

# Gerar embeddings
embedding_1 = embedder.encode(generated_text_1)
embedding_2 = embedder.encode(generated_text_2)
reference_embedding = embedder.encode(reference_text)

# Calcular similaridade
similarity_1 = cosine_similarity([embedding_1], [reference_embedding])
similarity_2 = cosine_similarity([embedding_2], [reference_embedding])

print("Cosine Similarity (Texto 1):", similarity_1[0][0])
print("Cosine Similarity (Texto 2):", similarity_2[0][0])
```
